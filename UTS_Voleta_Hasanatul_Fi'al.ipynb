## c. Apakah terdapat outlier pada data tersebut? Jelaskan apa itu outlier!

Adanya outlier berarti ada nilai yang berada jauh di luar kisaran nilai-nilai lain dalam dataset. Keberadaan outlier yang tidak ditangani dengan baik menimbulkan data yang menyimpang signifikan dari yang diharapkan.

Cek keberadaan outliers menggunakan boxplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Kolom numerik yang ingin dicek outliernya
# Only include numerical columns for outlier detection
num_cols = ['Age', 'SibSp', 'Parch', 'Fare']

# Visualisasi Boxplot
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i) # Adjusted subplot layout for 4 plots
    sns.boxplot(y=datatitanic[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```

Identifikasi outlier berdasarkan distribusi kuartil.

```python
def detect_outliers_iqr(datatitanic, column):
    Q1 = datatitanic[column].quantile(0.25)  # Kuartil 1
    Q3 = datatitanic[column].quantile(0.75)  # Kuartil 3
    IQR = Q3 - Q1  # Rentang interkuartil
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return datatitanic[(datatitanic[column] < lower_bound) | (datatitanic[column] > upper_bound)]

# Cek jumlah outlier untuk setiap kolom numerik
for col in num_cols:
    outliers = detect_outliers_iqr(datatitanic, col)
    print(f"{col}: {len(outliers)} outliers")
```

Hasil identifikasi tersebut menunjukkan bahwa kolom yang mengandung outlier yaitu kolom Age sebanyak 66 outliers, kolom SibSp sebanyak 46 outliers, kolom Parch sebanyak 213 outliers, dan kolom Fare sebanyak 116 outliers .

## d. Apakah outlier pada data tersebut perlu diatasi serta bagaimana cara mengatasinya?

Dampak keberadaan outlier terhadap hasil analisis yaitu menjadikan hasil analisis diambil dari data yang tidak valid atau bermasalah. Hal ini menimbulkan permasalahan juga dalam pengambilan keputusan berdasarkan hasil analisis tersebut.


Oleh karena itu, agar tidak menimbulkan permasalahan pada tahap analisis data, maka outliers pada data tersebutakan diatasi yaitu menggunakan metode winsorizing.

Winsorizing adalah salah satu teknik untuk menangani outlier dengan cara mengganti nilai ekstrem (outlier) dengan nilai yang lebih dekat ke distribusi data utama. Alasan mengapa winsorizing digunakan untuk menangani outlier adalah karena metode ini mengurangi pengaruh outlier tanpa menghapus data, menjaga distribusi, meningkatkan robustnes analisis statistik, serta memperbaiki hasil pemodelan. Winsorizing juga lebih fleksibel dibandingkan penghapusan outlier dan mudah diterapkan.

Outlier ditangangi dengan metode winsorizing.

```python
def winsorize_iqr(datatitanic, column):
    Q1 = datatitanic[column].quantile(0.25)
    Q3 = datatitanic[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    datatitanic[column] = np.where(datatitanic[column] < lower_bound, lower_bound, datatitanic[column])
    datatitanic[column] = np.where(datatitanic[column] > upper_bound, upper_bound, datatitanic[column])

# Terapkan Winsorizing
for col in num_cols:
    winsorize_iqr(datatitanic, col)
```

## a. Jelaskan proses pembagian data menjadi data training dan testing serta alasan mengapa perlu membagi data tersebut.

Pemisahan data training dan testing adalah keputusan penting dalam membangun model machine learning yang andal. Meskipun tidak ada aturan baku tentang proporsi pembagian data, terdapat praktik umum yang bisa dipertimbangkan:

Proses pembagian:
1.  Menentukan Target (y) dan Fitur (X)
2.  Menentukan Proporsi Pembagian
3.  Melakukan Pembagian Data Secara Acak
4.  Menggunakan Fungsi Pembagian Data, biasanya menggunakan fungsi seperti train_test_split() dari sklearn
5.  Validasi Pembagian

Alasan perlu membagi data menjad data training dna testing:
*   Menghindari Overfitting, yaitu kondisi di mana model terlalu menyesuaikan diri terhadap data pelatihan sehingga kehilangan kemampuan untuk melakukan generalisasi terhadap data baru. Dengan menggunakan data testing yang belum pernah dilihat model sebelumnya, kita dapat mengevaluasi sejauh mana model mampu bekerja pada data di luar data pelatihan.
*   Evaluasi Kinerja Secara Adil. Dengan adanya data testing, kita dapat melakukan evaluasi terhadap performa model secara lebih objektif dan adil. Evaluasi pada data yang berbeda dari data pelatihan memberikan gambaran yang lebih akurat mengenai kemampuan prediktif model di dunia nyata.












## b. Berapa persen pembagiannya? Mengapa memilih membagi dengan proporsi tersebut?

*   Pemisahan Standar: Pembagian yang sering digunakan adalah 70-30 atau 80-20, di mana 70% atau 80% data dialokasikan untuk training dan 30% atau 20% untuk testing. Ini merupakan titik awal yang baik untuk banyak kasus, termasuk dataset titanic.
*   Data yang Lebih Besar: Jika dataset sangat besar, pemisahan seperti 90-10 atau 95-5 bisa dipertimbangkan. Model dengan lebih banyak data training cenderung memiliki kinerja lebih baik, tetapi perlu diingat bahwa data testing yang cukup juga diperlukan untuk evaluasi yang akurat.

2) Model KNN

K-Nearest Neighbors (KNN) adalah metode klasifikasi berbasis instance (instance-based learning) yang mengklasifikasikan sebuah data baru berdasarkan mayoritas kelas dari K tetangga terdekatnya dalam ruang fitur.

Langkah-langkah Pemodelan:

1. Pencarian Parameter Optimal: melakukan Grid Search untuk mencari parameter optimal untuk model KNN (walaupun kode Anda menggunakan MLPClassifier, prinsipnya sama).
2. Membuat dan Melatih Model

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.impute import SimpleImputer # Import SimpleImputer

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)

# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Impute missing values using SimpleImputer before fitting
imputer = SimpleImputer(strategy='mean') # Create an imputer instance
X = imputer.fit_transform(X) # Fit and transform the data

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

3) Model Neural Network

Neural Network (Jaringan Saraf Tiruan) adalah model pembelajaran mesin yang terinspirasi oleh struktur jaringan saraf biologis. Neural Network terdiri dari sekumpulan node (neuron) yang terhubung dalam lapisan-lapisan tertentu: input layer, hidden layer(s), dan output layer.

Langkah kerja:
1. Input data masuk ke input layer.
2. Data diproses melalui hidden layers (menggunakan kombinasi linear + fungsi aktivasi).
3. Output dihasilkan di output layer.
4. Loss dihitung untuk mengukur kesalahan.
5. Backpropagation dilakukan untuk memperbarui bobot dan bias.
6. Proses diulang hingga model mencapai performa yang optimal

```python
#Menguji parameter yang paling optimum untuk model

from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.model_selection import StratifiedKFold

cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=100)

# Tentukan parameter yang ingin diuji
parameters = {
    'hidden_layer_sizes': [(50,), (100,), (50, 30), (100, 50)],
    'learning_rate_init': [0.001, 0.01, 0.1],
    'max_iter': [100, 200, 300]
}

# Buat model MLPClassifier
clf = MLPClassifier(random_state=100)

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, parameters, cv=cv_strategy)
# cv merupakan jumlah lipatan validasi silang (cross-validation)

# Lakukan pencarian grid pada data training
grid_search.fit(X, y)
```

```python
# Tampilkan parameter terbaik
print("Parameter terbaik:", grid_search.best_params_)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Import library yang diperlukan
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

# Asumsikan X_train, X_test, y_train, y_test sudah ada dari proses sebelumnya

# 1. NAIVE BAYES
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)

# Evaluasi Naive Bayes
print("=== Evaluasi Naive Bayes ===")
print("Accuracy :", accuracy_score(y_test, y_pred_nb))
print("Precision:", precision_score(y_test, y_pred_nb))
print("Recall   :", recall_score(y_test, y_pred_nb))
print("F1-Score :", f1_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))

# 2. K-NEAREST NEIGHBORS (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)  # kamu bisa ganti k kalau mau
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

# Evaluasi KNN
print("\n=== Evaluasi KNN ===")
print("Accuracy :", accuracy_score(y_test, y_pred_knn))
print("Precision:", precision_score(y_test, y_pred_knn))
print("Recall   :", recall_score(y_test, y_pred_knn))
print("F1-Score :", f1_score(y_test, y_pred_knn))
print(classification_report(y_test, y_pred_knn))

# NEURAL NETWORK (MLPClassifier)
nn_model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42)
nn_model.fit(X_train, y_train)
y_pred_nn = nn_model.predict(X_test)

# Evaluasi Neural Network
print("\n=== Evaluasi Neural Network ===")
print("Accuracy :", accuracy_score(y_test, y_pred_nn))
print("Precision:", precision_score(y_test, y_pred_nn))
print("Recall   :", recall_score(y_test, y_pred_nn))
print("F1-Score :", f1_score(y_test, y_pred_nn))
print(classification_report(y_test, y_pred_nn))
```

Nilai matrik evaluasi masing-masing model menunjukkan
*   Accuracy = Seberapa banyak prediksi yang benar secara umum.
*   Precision = Dari yang diprediksi positif, seberapa banyak yang benar.
*   Recall = Dari yang benar-benar positif, berapa banyak yang berhasil ditemukan.
*   F1-Score = Gabungan rata-rata precision dan recall dalam satu angka.
